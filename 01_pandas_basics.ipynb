```python
import pandas as pd 
print ("Pandas Version", pd.__version__)
```

```python
data = {

    "Employee": ["Alice","Bob","Charlie","Diana"],
    "Department" : ["Finance","HR","IT","Finance"],
    "Salary" : [5000,2000,8000,25000],
    "Role" :["Senior","Assistant","Manager","Senior Manager"]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Average Salary
print("Average Salary", df["Salary"].mean())

#Filter finance department only
finance = df[df["Department"] == "Finance"]
print("\nFinance Team:")
print(finance)
```

```python
#Display Role and salary
print("Role", df["Role"])

#Filter by Role Manager
manager=df[df["Role"] == "Manager"]
print("\nRole Manage:")
print(manager)
```

```python
#sum the salary
print("Sum of Salary:",df["Salary"].sum())
```
